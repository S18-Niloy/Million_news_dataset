```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM
```

```python
# Load the dataset
df = pd.read_csv('/kaggle/input/million-headlines/abcnews-date-text.csv')
```

```python
# Preprocessing
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['headline_text'])
vocab_size = len(tokenizer.word_index) + 1
```

```python
# Convert text to sequences
sequences = tokenizer.texts_to_sequences(df['headline_text'])
```

```python
# Pad sequences
max_len = max(len(seq) for seq in sequences)
sequences = pad_sequences(sequences, maxlen=max_len)
```

```python
# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(sequences, df['publish_date'], test_size=0.2, random_state=42)
```

```python
# Build the RNN model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=100, input_length=max_len))
model.add(LSTM(units=64))
model.add(Dense(units=1))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(X_train, y_train, batch_size=64, epochs=10)
```

```python
# Evaluate the model on the test set
loss = model.evaluate(X_test, y_test)
print('Test Loss:', loss)
```

```python
# Make predictions on the test set
predictions = model.predict(X_test)
```

```python
# Convert predictions to binary values
predictions_binary = [1 if pred >= 0.5 else 0 for pred in predictions]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
```

```python
#confusion matrix
cm = confusion_matrix(y_test, predictions_binary)
print(cm)
```
